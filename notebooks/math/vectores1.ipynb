## VECTORES EN EL PLANO

**Definición de vector en el plano**  
*Un **vector en el plano** es un par ordenado de números reales (x, y). Los números x y y son los **componentes** del vector (x, y).*  


```python
import matplotlib.pyplot as plt
import numpy as np

#Coordenadas del origen del vector
origen_x= 0
origen_y= 0

#Componentes del vector (dx, dy)
vector_x= 3
vector_y= 4

#Dibujar el vector
plt.figure(figsize=(5,5))
plt.quiver(origen_x, origen_y, vector_x, vector_y, angles='xy', scale_units='xy', scale= 1, color='r')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector simple en 2D')
plt.show()
```

**Definición de magnitud y dirección de un vector**  
La **magnitud** de un vector **A** denotado por $|\vec{A}|$, es la longitud de cualquiera de sus representaciones, y la **dirección** del vector cero en la dirección de cualquiera de sus representaciones.  

**Teorema**  
Si **A** es el vector $(a_1,a_2)$, entonces $|\vec{A}|=\sqrt{a_¡^2 + a_2^2}$

El **ángulo director** de cualquier vector diferente del vector cero es el ángulo $\theta$ medido desde la parte positiva del eje x en el sentido contrario a las manecillas del reloj hasta la representación de posición del vector.  
Si $\theta$ se mide en radianes, entonces $0\le \theta \le 2\pi$.  
Si **A** = $(a_1, a_2)$ entonces; $\tan \theta =\frac{a_2}{a_1}$ Si $a_1 \neq 0$

```python
import matplotlib.pyplot as plt
import numpy as np

#Coordenadas del origen del vector
origen_x = 0
origen_y = 0

#Componentes del vector (x, y)
vector_x = -1
vector_y = 1
angulo = np.arctan2(vector_y, vector_x)

# Puntos del ángulo
r= 0.5
theta = np.linspace(0, angulo, 50)

# Coordenadas del arco
arco_x = r*np.cos(theta)
arco_y = r*np.sin(theta)

#Dibujo del vector
plt.figure(figsize=(5,5))
plt.quiver(origen_x, origen_y, vector_x, vector_y, angles ='xy', scale_units ='xy', scale = 1, color = 'r')
plt.grid(True)
plt.xlim(-1.5, 1)
plt.ylim(-1, 1.5)
plt.axhline(0, color = 'black', linewidth = 1) # Eje x
plt.axvline(0, color = 'black', linewidth = 1) # Eje y
# Dibujo del arco
plt.plot(arco_x, arco_y, color = 'blue', linewidth = 1, ) 
plt.show
```
