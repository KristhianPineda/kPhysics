# Vectores, rectas, planos y superficies en el espacio

## VECTORES EN EL PLANO

**Definición de vector en el plano**  
Un **vector en el plano** es un par ordenado de números reales (x, y). Los números x y y son los **componentes** del vector (x, y).  


**Definición de magnitud y dirección de un vector**  
La **magnitud** de un vector **A** denotado por $|\vec{A}|$, es la longitud de cualquiera de sus representaciones, y la **dirección** del vector cero en la dirección de cualquiera de sus representaciones.  

**Teorema**  
Si **A** es el vector $(a_1,a_2)$, entonces $|\vec{A}|=\sqrt{a_1^2 + a_2^2}$

```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas del vector
vector_x = -4
vector_y = 5

# Dibujar vectores
plt.figure(figsize=(5,5))
plt.quiver(6, -2, vector_x, vector_y, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, vector_x, vector_y, angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector simple en 2D')
plt.axhline(0, color = 'black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.show()

# Calcular magnitud del vector
angulo=np.arctan2(vector_y, vector_x)
magnitud = np.sqrt((vector_x)**2 + (vector_y)**2)
print("La magnitud es = ", f"{magnitud:.2f}")
angulo_deg = np.degrees(angulo)
print(f"{angulo_deg:.2f}° grados")
```

Si:  
x = coordenada inicial en el eje x,  
y = coordenada inicial en el eje y,  
u = componente x del vector,  
v = componente y del vector,  
$x_1$ = coordenada terminal en x,  
$y_1$ = coordenada terminal en y; entonces; $$ x_1=u+x$$ $$y_1=v+y$$

**Ejemplo 1** Sean **A** el vector (-4. 5) y **P** el punto (6, -2), dibuje la representación de posición de **A** y también la representación particular de **A** que tiene a **P** como su punto inicial.  

**Datos relevantes**  
$u=-4$  
$v=5$  
$x=6$  
$y=-2$  
**Estrategia**
$$x_1=u+x$$ $$y_1=v+y$$  
**Aplicación de estrategia**
$$x_1=-4+6$$ $$x_1=2$$ $$y_1=5-2$$ $$y_1=3$$
**Resultado**  
posición inicial: $(6, 2)$, posición terminal: $(2, 3)$ 

El **ángulo director** de cualquier vector diferente del vector cero es el ángulo $\theta$ medido desde la parte positiva del eje x en el sentido contrario a las manecillas del reloj hasta la representación de posición del vector.  
Si $\theta$ se mide en radianes, entonces $0\le \theta \le 2\pi$.  
Si **A** = $(a_1, a_2)$ entonces; $\tan \theta =\frac{a_2}{a_1}$ Si $a_1 \neq 0$

```python
import matplotlib.pyplot as plt
import numpy as np

#Coordenadas del origen del vector
origen_x = 0
origen_y = 0

#Componentes del vector (x, y)
vector_x = -1
vector_y = 1
angulo = np.arctan2(vector_y, vector_x)

# Puntos del ángulo
r= 0.5
theta = np.linspace(0, angulo, 50)

# Coordenadas del arco
arco_x = r*np.cos(theta)
arco_y = r*np.sin(theta)

#Dibujo del vector
plt.figure(figsize=(5,5))
plt.quiver(origen_x, origen_y, vector_x, vector_y, angles ='xy', scale_units ='xy', scale = 1, color = 'r')
plt.grid(True)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color = 'black', linewidth = 1) # Eje x
plt.axvline(0, color = 'black', linewidth = 1) # Eje y
plt.title("Angulo director en 2D")

# Dibujo del arco
plt.plot(arco_x, arco_y, color = 'blue', linewidth = 1) 
plt.show

# Calcular el ángulo de dirección
print("La dirección es=", f"{angulo:.2f}")
angulo_deg = np.degrees(angulo)
print(f"{angulo_deg:.2f}° grados")
```

Observe que si **A** = $(a_1, a_2)$ y $\theta$ es el ángulo director de **A** entonces;  
$a_1$ = | **A** | $\cos\theta$ y $a_2$ = | **A** | $\sin\theta$

**Ejemplo 2** Suponga que **P** es el punto (-1, 8) y **Q** es el punto (3, 2), determine el vector **A** que tiene $\vec{PQ}$

**Datos relevantes**  
$x_1=3$  
$x=-1$  
$y_1=2$  
$y=8$  
**Estrategia**
$$x_1=u+x$$ $$y_1=v+y$$  
**Aplicación de estrategia**

**Definición de suma de vectores**  
La **suma** de los vectores **A** = $(a_1,a_2)$ y **B** = $(b_1, b_2)$ es el vector **A** + **B** definido por: $$A + B = (a_1 + b_1, a_2 + b_2)$$ 

**Ejemplo ilustrativo** Si **A** = (3, -1) y **B** = (-4, 5), entonces; $$A+B=(3+(-4), (-1)+5)$$ $$A+B=(-1, 4)$$

```python
import matplotlib.pyplot as plt

# Componentes de los vectores
vector_ax=3
vector_ay=-1
vector_bx=-4
vector_by=5
resultante_x=-1
resultante_y=4
x_values=[3,-1]
y_values=[-1, 4]
x1_values=[-4, -1]
y1_values=[5, 4]

# Plano xy
plt.figure(figsize=(7, 7))
plt.grid(True)
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Dibujo de vectores
plt.quiver(0, 0, vector_ax, vector_ay, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, vector_bx, vector_by, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, resultante_x, resultante_y, angles='xy', scale_units='xy', scale=1, color='green')
plt.plot(x_values, y_values, color='magenta', linestyle='-.', linewidth=1)
plt.plot(x1_values, y1_values, color='magenta', linestyle='--', linewidth=1)
plt.show
```

**Ejemplo 3** Dos fuerzas de 200 lb y 250 lb forman un ángulo de $\frac{1}{3}$ $\pi$ entre si y están aplicadas a un objeto en el mismo punto. **(a)** Determine la intensidad resultante, y **(b)** el ángulo que forma la resultante con la fuerza de 200 lb.  

```python
import matplotlib.pyplot as plt
import numpy as np

# Componentes del vector resultante
resultante_x=325
resultante_y=216.5

# Ángulo
angulo=np.arctan2(resultante_y, resultante_x)

# Puntos del ángulo
r=100
theta=np.linspace(0, angulo, 50)

# Coordenadas del arco
arc_x=r*np.cos(theta)
arc_y=r*np.sin(theta)

# Creación del plano xy
plt.figure(figsize=(7, 7))
plt.xlim(-20, 400)
plt.ylim(-20, 400)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.title("Diagrama de cuerpo libre")
plt.show

# dibujo de los vectores que representan las fuerzas
plt.quiver(0, 0, 200, 0, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, 125, 216.5, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, resultante_x, resultante_y, angles='xy', scale_units='xy', scale=1, color='green')

# Dibujo de arco
plt.plot(arc_x, arc_y, color='blue', linewidth=1)
```

**Datos relevantes**  
$F_1=200 lb$  
$F_2=250 lb$  
$\theta =\frac{1}{3}\pi$  
**Estrategia**  
$$A+B$$ $$|A|\cos\theta$$ $$|A|\sin\theta$$ $$|B|\cos\theta$$ $$|B|\sin\theta$$ $$||A+B||=\sqrt{(a_1+b_1)^2+(a_2-b_2)^2}$$  
**Aplicación de la estrategia**  
$$a_1=200\cos0°$$ $$a_1=200$$ $$a_2=200\sin0°$$ $$a_2=0$$ $$b_1=250\cos\frac{1}{3}\pi$$ $$b_1=125$$ $$b_2=250\sin\frac{1}{3}\pi$$ $$b_2=216.5$$ $$|A+B|=(325, 216.5)$$ $$|A+B|=\sqrt{(325)^2+(216.5)^2}$$ $$|A+B|=390.5$$ $$\tan\theta=\frac{216.5}{325}$$ $$\theta=0.5877$$  
**Resultado**  
La intensidad resultante es 390.5 y el ángulo que forma la resultante con la fuerza de 200 lb es 0.5877
