# Vectores, rectas, planos y superficies en el espacio

## VECTORES EN EL PLANO

**Definición de vector en el plano**  
Un **vector en el plano** es un par ordenado de números reales (x, y). Los números x y y son los **componentes** del vector (x, y).  


**Definición de magnitud y dirección de un vector**  
La **magnitud** de un vector **A** denotado por $|\vec{A}|$, es la longitud de cualquiera de sus representaciones, y la **dirección** del vector cero en la dirección de cualquiera de sus representaciones.  

**Teorema**  
Si **A** es el vector $(a_1,a_2)$, entonces $|\vec{A}|=\sqrt{a_1^2 + a_2^2}$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from IPython.display import display

permitido = {"__builtins__": None, "sqrt": np.sqrt, "pi":np.pi, "e": np.e}

# Ingresa los componentes del vector
print("Ingresa los componentes del vector")
vec_x_str = input("Componente x: ")
vec_y_str = input("Componente y: ")
origen_x_str = input("Origen x: ")
origen_y_str = input("Origen y: ")
vec_x = eval(vec_x_str, permitido)
vec_y = eval(vec_y_str, permitido)
origen_x = eval(origen_x_str, permitido)
origen_y = eval(origen_y_str, permitido)

# Ingresa los parámetros de los límites de la gráfica
lim_x=abs(vec_x) + 1
lim_y=abs(vec_y) + 1

# Dibuja el plano
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_aspect=('equal')
ax.set_xlim(-lim_x, lim_x)
ax.set_ylim(-lim_y, lim_y)
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Trazo de un vector simple en 2D")
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Dibuja los vectores
ax.quiver(origen_x, origen_y, vec_x, vec_y, angles='xy', scale_units='xy', scale=1, color='r')
ax.text(lim_x/2, lim_y-2, r"$\vec{A}$", fontsize=14, color='r')
plt.show()

# Calcula la magnitud del vector
theta1=np.arctan2(vec_y, vec_x)
magnitud = np.sqrt((vec_x)**2 + (vec_y)**2)
angulo_deg = np.degrees(theta1)

# Muestra el calculo de la magnitud y angulo del vector
vec_x, vec_y = sp.symbols('v_x v_y', real = True)
v = sp.Matrix([vec_x, vec_y])
mod_v = sp.sqrt(v.dot(v))
display(sp.Eq(sp.Symbol(r'|\vec{v}|'), mod_v))
display(sp.Eq(sp.Symbol(r'|\vec{v}|'), magnitud))
theta=sp.atan2(vec_y, vec_x)
display(sp.Eq(sp.Symbol(r'\theta_radianes'), theta))
display(sp.Eq(sp.Symbol(r'\theta_grados'), theta1))
```

Si:  
x = coordenada inicial en el eje x,  
y = coordenada inicial en el eje y,  
u = componente x del vector,  
v = componente y del vector,  
$x_1$ = coordenada terminal en x,  
$y_1$ = coordenada terminal en y; entonces; $$ x_1=u+x$$ $$y_1=v+y$$

El **ángulo director** de cualquier vector diferente del vector cero es el ángulo $\theta$ medido desde la parte positiva del eje x en el sentido contrario a las manecillas del reloj hasta la representación de posición del vector.  
Si $\theta$ se mide en radianes, entonces $0\le \theta \le 2\pi$.  
Si **A** = $(a_1, a_2)$ entonces; $\tan \theta =\frac{a_2}{a_1}$ Si $a_1 \neq 0$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from IPython.display import display 

permitido = {"__builtins__": None, "sqrt": np.sqrt, "pi":np.pi, "e": np.e}

# Ingresa los componentes del vector
print("Ingresa los componentes del vector")
vec_x_str = input("Componente x: ")
vec_y_str = input("Componente y: ")
origen_x_str = input("Origen x: ")
origen_y_str = input("Origen y: ")
vec_x = eval(vec_x_str, permitido)
vec_y = eval(vec_y_str, permitido)
origen_x = eval(origen_x_str, permitido)
origen_y = eval(origen_y_str, permitido)
angulo = np.arctan2(vec_y, vec_x)

# Normaliza
if angulo < 0:
    angulo += 2*np.pi

# Ingresa los parametros de los límites de la gráfica
lim_x=abs(vec_x) + abs(origen_x) + 1
lim_y=abs(vec_y) + abs(origen_y) + 1

# Puntos del ángulo
r= 0.5*np.hypot(vec_x, vec_y)
theta = np.linspace(0, angulo, 200)

# Coordenadas del arco
arco_x = origen_x + r*np.cos(theta)
arco_y = origen_y + r*np.sin(theta)
linea_x = [origen_x, origen_x + r]
linea_y = [origen_y, origen_y]

# Dibuja el plano
fig, ax = plt.subplots(figsize=(7,7))
ax.set_aspect('equal')
ax.grid(True)
ax.set_xlim(-lim_x, lim_x)
ax.set_ylim(-lim_y, lim_y)
ax.axhline(0, color = 'black', linewidth = 1) # Eje x
ax.axvline(0, color = 'black', linewidth = 1) # Eje y
ax.set_title("Angulo director en 2D")


# Dibuja el vector
ax.quiver(origen_x, origen_y, vec_x, vec_y, angles ='xy', scale_units ='xy', scale = 1, color = 'r')

# Dibujo del arco
ax.plot(arco_x, arco_y, color = 'blue', linewidth = 1)
ax.plot(linea_x, linea_y, color = 'green', linewidth = 2, linestyle = '--')
plt.show()

# Muestra el ángulos de dirección en radianes y grados
angulo_deg = np.degrees(angulo)
magnitud = np.sqrt((vec_x)**2 + (vec_y)**2)

# Muestra el calculo de la magnitud y angulo del vector
vec_x, vec_y = sp.symbols('v_x v_y', real = True)
v = sp.Matrix([vec_x, vec_y])
mod_v = sp.sqrt(v.dot(v))
display(sp.Eq(sp.Symbol(r'|\vec{v}|'), mod_v))
display(sp.Eq(sp.Symbol(r'|\vec{v}|'), magnitud))
theta=sp.atan2(vec_y, vec_x)
display(sp.Eq(sp.Symbol(r'\theta_rad'), angulo))
display(sp.Eq(sp.Symbol(r'\theta_grados'), angulo_deg))
```

Observe que si **A** = $(a_1, a_2)$ y $\theta$ es el ángulo director de **A** entonces;  
$a_1$ = | **A** | $\cos\theta$ y $a_2$ = | **A** | $\sin\theta$

**Ejemplo 3** Dos fuerzas de 200 lb y 250 lb forman un ángulo de $\frac{1}{3}$ $\pi$ entre si y están aplicadas a un objeto en el mismo punto. **(a)** Determine la intensidad resultante, y **(b)** el ángulo que forma la resultante con la fuerza de 200 lb.  

```python
import matplotlib.pyplot as plt
import numpy as np

# Componentes del vector resultante
resultante_x=325
resultante_y=216.5

# Ángulo
angulo=np.arctan2(resultante_y, resultante_x)

# Puntos del ángulo
r=100
theta=np.linspace(0, angulo, 50)

# Coordenadas del arco
arc_x=r*np.cos(theta)
arc_y=r*np.sin(theta)

# Creación del plano xy
plt.figure(figsize=(7, 7))
plt.xlim(-20, 400)
plt.ylim(-20, 400)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.title("Diagrama de cuerpo libre")
plt.show

# dibujo de los vectores que representan las fuerzas
plt.quiver(0, 0, 200, 0, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, 125, 216.5, angles='xy', scale_units='xy', scale=1, color='blue')
plt.quiver(0, 0, resultante_x, resultante_y, angles='xy', scale_units='xy', scale=1, color='green')

# Dibujo de arco
plt.plot(arc_x, arc_y, color='blue', linewidth=1)
```

**Datos relevantes**  
$F_1=200 lb$  
$F_2=250 lb$  
$\theta =\frac{1}{3}\pi$  
**Estrategia**  
$$A+B$$ $$|A|\cos\theta$$ $$|A|\sin\theta$$ $$|B|\cos\theta$$ $$|B|\sin\theta$$ $$||A+B||=\sqrt{(a_1+b_1)^2+(a_2-b_2)^2}$$  
**Aplicación de la estrategia**  
$$a_1=200\cos0°$$ $$a_1=200$$ $$a_2=200\sin0°$$ $$a_2=0$$ $$b_1=250\cos\frac{1}{3}\pi$$ $$b_1=125$$ $$b_2=250\sin\frac{1}{3}\pi$$ $$b_2=216.5$$ $$|A+B|=(325, 216.5)$$ $$|A+B|=\sqrt{(325)^2+(216.5)^2}$$ $$|A+B|=390.5$$ $$\tan\theta=\frac{216.5}{325}$$ $$\theta=0.5877$$  
**Resultado**  
La intensidad resultante es 390.5 y el ángulo que forma la resultante con la fuerza de 200 lb es 0.5877

**Ejemplo ilustrativo** Si $\vec{A}=(4, -2)$ y $\vec{B}=(6. -3)$ entonces;  

$$\vec{A}-\vec{B}=(4, 2) - (6, -3)$$ $$=(4, -2) + (-6, 3)$$ $$=(-2, 1)$$

**Ejemplo 4** Un avión puede volar a 300 mi/h. Si el viento sopla hacia el este a 50 mi/h, ¿Cuál debe ser el enfilamiento del avión para que el curso sea de 30°? ¿Cuál será la velocidad a tierra del avión si vuela en este curso?  

**Datos relevantes**  
$\vec{A}$= velocidad del avión  
|$\vec{A}-\vec{B}$|= 300 mi/h (velocidad del avión en el aire)  
$\vec{B}$= 50 mi/h (velocidad del viento)  
$\phi$= ángulo de desviación  
$\theta$= ángulo director  
**Estrategia**  
$$\frac{\sin{\phi}}{\vec{B}}=\frac{\sin{60°}}{|\vec{A}-\vec{B}|}$$ $$\frac{\sin{180°-\theta}}{|vec{A}|}=\frac{\sin{60°}}{|\vec{A}-\vec{B}|}$$  
**Aplicación de la estrategia**  
$$\sin{\phi}=\frac{(50)(\sin{60°)}}{300}$$ $$\phi=8.3°$$ por tanto; $\theta$ = 60°+8.3°=68.3° $$|\vec{A}|=\frac{(300)(\sin(111.7°)}{\sin{60°}}$$ $$|\vec{A}|=322$$  
**Resultado**
El enfilamiento del avión debe ser 90°-$\theta$=21.7°, y si el avión vuela en este curso su velocidad a tierra será de 322 mi/h.

**Definición del producto de un escalar y un vector**  
Si *c* es un escalar y **A** es un vector $(a_1,a_2)$, entonces el **producto** de *c* y **A**, denotado por *c***A**, es el vector definido por: $$\textit{c}\vec{A}=\textit{c}(a_1,a_2)$$ $$=(\textit{c}a_1, \textit{c}a_2)$$

**Ejemplo ilustrativo** Si **A** =(4, -5), entonces; $$3A=3(4, -5)$$ $$=(12, -15)$$ 

**Definición de espacio vectorial real**  
Un **espacio vectorial real** $\textit{V}$ es un conjunto de elementos, llamados $\textit{vectores}$, junto con el conjunto de números reales, denominados *escalares*, con dos operaciones llamadas *adición vectorial y multiplicación por un escalar,* tal que para cada par de vectores $\vec{A}$ y $\vec{B}$ en *V* y para cualquier escalar *c*, se definen los vectores **A** + **B** y *c***A** de modo que las propiedades se cumplan.    

**Vectores unitarios**  
$(a_1, a_2)$= $a_1\hat{\imath}$ + $a_2\hat{\jmath}$ 

Se dice que $\hat{\imath}$ y $\hat{\jmath}$ forman una base para el espacio vectorial, el número de elementos de una base de un espacio vectorial se denomina **dimensión** del espacio vectorial. 

$$\textit{A}=||A||(\cos\theta\hat{\imath}+\sin\theta\hat{\jmath})$$

**Teorema**  
Si el vector **A** =$a_1\hat{\imath}$ + $a_2\hat{\jmath}$ es diferente del vector cero, entonces el vector unitario **U** que tiene la misma dirección y el mismo sentido de **A** esta definido por $$U=\frac{a_1}{||A||}\hat{\imath} + \frac{a_2}{||A||}\hat{\jmath}$$

**Ejemplo 5** Dados **A**=$3\hat{\imath}$ + $\hat{\jmath}$ y **B** = $-2\hat{\imath}+ 4\hat{\jmath}$, obtenga el vector unitarioque tiene la misma dirección de **A** - **B**   

**Solución:**  
$$A - B =(3\hat{\imath} + \hat{\jmath}) - (-2\hat{\imath} + 4\hat{\jmath})$$ $$=5\hat{\imath} - 3\hat{\jmath}$$ $$||A-B||=\sqrt{(5)^2+(-3)^2}$$ $$=\sqrt{34}$$ $$U=\frac{5}{\sqrt{34}}\hat{\imath} - \frac{3}{\sqrt{34}}\hat{\jmath}$$
